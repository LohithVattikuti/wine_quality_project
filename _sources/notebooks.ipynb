# Predicting Wine Quality - A Machine Learning Approach

This project focuses on predicting the quality of wine using machine learning models based on chemical properties.
The goal is to build a reliable model and deploy it for real-time predictions through a web application.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (ensure the file winequality-red.csv is in the specified path)
try:
    df = pd.read_csv("/Users/lohithvattikuti/Desktop/Wine_Quality_project/data/winequality-red.csv", delimiter=';')  # Correct path with delimiter
    print("Dataset loaded successfully!")
except FileNotFoundError as e:
    print(f"Error: {e}")
    print("Please ensure the winequality-red.csv file exists at the specified path.")

# Display the first few rows
if 'df' in locals():
    print(df.head())

    # Basic info about the dataset
    print("\nDataset Information:")
    print(df.info())

    # Check for missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Summary statistics
    print("\nSummary Statistics:")
    print(df.describe())

    # Visualize the distribution of wine quality
   
    plt.figure(figsize=(8, 6))
    sns.countplot(x='quality', data=df, hue='quality', palette='viridis', dodge=False, legend=False)
    plt.title('Distribution of Wine Quality')
    plt.xlabel('Wine Quality')
    plt.ylabel('Count')
    plt.show()

    # Correlation matrix
    plt.figure(figsize=(10, 8))
    corr = df.corr()
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Matrix')
    plt.show()

    # Pair plot for selected features
    selected_features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']
    sns.pairplot(df[selected_features], hue='quality', palette='viridis')
    plt.show()

    # Observations
    print("\nObservations:")
    print("""
    1. The distribution of wine quality shows that most wines fall into a specific range (e.g., 5-7).
    2. The correlation matrix shows relationships between features. Alcohol and quality seem positively correlated.
    3. Volatile acidity appears to negatively correlate with wine quality.
    4. Further exploration and feature engineering may be needed to improve predictions.
    """)
else:
    print("Dataset could not be loaded. Please resolve the issue and try again.")
```
